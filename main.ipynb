```python
# signifies game is active
game = 1

player_hand = functions.deal_cards(functions.deck)
dealer_hand = functions.deal_cards(functions.deck)

player_card1 = str(player_hand[0])
player_card2 = str(player_hand[1])
dealer_card1 = str(dealer_hand[0])
dealer_card2 = str(dealer_hand[1])

print("BLACKJACK GAME!\n")
print("You were dealt: " + player_card1 + "," + player_card2)
print("Your total is: " + str(functions.calc_total(player_hand)))
print("The dealer shows one card: " + dealer_card1)

# you win if dealt a blackjack at the start
if(functions.calc_total(player_hand) == 21):
    print("You win! Blackjack!")
    game = 0

#dealer wins if dealt blackjack at start
elif(functions.calc_total(dealer_hand) == 21):
    print("You lost! The dealer got a blackjack")
    game = 0

# signifies game is active
while(game == 1):
    # handles inputs 'hit', 'stand', 'quit'
    player = str(input("\nDo you want to 'Hit', 'Stand', 'Quit' "))
    
    # deals a new card to the player if he/she chooses to 'hit'
    if(player == "hit" or player == "Hit"):
        functions.hit(player_hand)
        print("You were dealt a: " + str(player_hand[-1]))
        print("Your new total is: " + str(functions.calc_total(player_hand)))
        
        player_total = functions.calc_total(player_hand)
        
        # if player total is greater than 21, dealer wins
        if(player_total > 21):
            print("Sorry you lost, you went over 21!")
            break
            game = 0
        # if player achieves 21 he wins by getting a blackjack
        elif(player_total == 21):
            print("You win! Blackjack!")
            break
            game = 0
                  
        # dealer's strategy will hit everytime if his total is less than 16
        while(functions.calc_total(dealer_hand) < 17):
            functions.hit(dealer_hand)
            
        dealer_total = functions.calc_total(dealer_hand)
        
        # if dealer total is greater than 21, player wins
        if(dealer_total > 21):
            print("You win! the dealer went over 21!")
            print("Their total was: " + str(dealer_total))
            game = 0
        # if dealer total is 21 they win by getting a blackjack
        elif(dealer_total == 21):
            print("You lost! the dealer got a blackjack")
            game = 0
        
    # if a player decides to confirm his total number to go agaisnt dealer
    elif(player == "stand" or player == "Stand"):
        
        # dealer's strategy will hit everytime if his total is less than 16
        while(functions.calc_total(dealer_hand) < 17):
            functions.hit(dealer_hand)
        
        player_total = functions.calc_total(player_hand)
        dealer_total = functions.calc_total(dealer_hand)
        
        # if dealer total is greater than 21, player wins
        if(dealer_total > 21):
            print("You win! the dealer went over 21!")
            print("Their total was: " + str(dealer_total))
            game = 0
         
        # if dealer total is greater than player and less than 22 dealer wins
        elif(dealer_total > player_total):
            print("You lose! the dealer's total is: " + str(dealer_total))
            print("Which is higher than yours: " + str(player_total))
            game = 0
        # if player total is greater than dealer and less than 22, player wins
        elif(dealer_total < player_total):
            print("You win! your total is " + str(player_total))
            print("Which is higher than the dealers " + str(dealer_total))
            game = 0

    # quits the game
    elif(player == "quit" or player == "Quit"):
        print("\nThanks for playing come back soon!")
        game = 0
    
    # if a user types an invalid input    
    else:
        print("Please select a valid option")
```
